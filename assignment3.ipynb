```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gdp = pd.read_csv('RealGDP17092022.csv')
gdp.head()
```

```python
gdp.reset_index(inplace=True)
```

```python
gdp.rename({'index':'Year', 'tyear':'Period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Livestock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MininingQua', 'MiningAndQuarrying':'CrudeOil',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'Minerals',
       'QuarryingAndOtherMinerals':'Manufacturing','Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBev',
       'FoodBeverageAndTobacco':'Textile', 'TextileApparelAndFootwear':'WoodProduct',
       'WoodAndWoodProducts':'PaperProduct', 'PulpPaperAndPaperProducts':'PhamaProduct',
       'ChemicalAndPharmaceuticalProducts':'NonMetalPrd', 'NonMetallicProducts':'PlasticProduct',
       'PlasticAndRubberProducts':'EEE', 'ElectricalAndElectronics':'IronSteel',
       'BasicMetalIronAndSteel':'VehiclesASS', 'MotorVehiclesAndAssembly':'OtherManu',
       'OtherManufacturing':'Gasstem', 'ElectricityGasSteamAndAirCon':'Watersupply',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'Accommodation',
       'AccommodationAndFoodServices':'Transport', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'Rail', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'TransService', 'TransportServices':'PostServices', 'PostAndCourierServices':'ICT',
       'InformationAndCommunication':'Publishing', 'Publishing':'MusicProduction',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'Entertainment',
       'ArtsEntertainmentAndRecreation':'FinInsurance', 'FinanceAndInsurance':'FinInstitution',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'SciTech',
       'ProfessionalScientificAndTechnicalServices':'Adminstration',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdminstration',
       'PublicAdministration':'Education', 'Education':'HealthService', 'HumanHealthAndSocialServices':'OtherService',
       'OtherServices':'GDP2010', 'GDP at 2010 Constant Basic Prices':'Nettaxs',
       'NetTaxesOnProducts':'GDP20102', 'GDP at 2010 Constant Market Prices':'NILL'},axis = 1,inplace =True)
gdp.head()
```

```python
gdp.drop('NILL',axis=1,inplace=True)
gdp.head()
```

```python
gdp = gdp[gdp['Period'] == 'Annual']
```

```python
fig,axes = plt.subplots(2,3,figsize =(15,8))
axes[0][0].bar(gdp['Year'],gdp['Agriculture'],color = 'magenta')
axes[0][0].set_xlabel('Year',color = 'green',fontsize =15)
axes[0][0].set_ylabel('GDP',color = 'green',fontsize =15)
axes[0][0].set_title('AGRICULTURE GDP IN MILLONS',color = 'blue',fontsize =15)

for label in axes[0][0].get_xticklabels():
    label.set_color('crimson')
for label in axes [0][0].get_xticklabels():
    label.set_color('crimson')

axes[0][1].step(gdp['Year'],gdp['ICT'],color = 'cyan')
axes[0][1].set_xlabel('Year',color = 'green',fontsize =15)
axes[0][1].set_ylabel('GDP',color = 'green',fontsize =15)
axes[0][1].set_title('ICT GDP IN MILLONS',color = 'blue',fontsize =15)

for label in axes[0][1].get_xticklabels():
    label.set_color('crimson')
for label in axes [0][1].get_xticklabels():
    label.set_color('crimson') 

    plt.suptitle('VARIOUS NIGERIA INDUSTRIES GDP DATA',color='blue',fontsize = 30) 

axes[0][2].bar(gdp['Year'],gdp['Education'],color = 'cyan')
axes[0][2].set_xlabel('Year',color = 'green',fontsize =15)
axes[0][2].set_ylabel('GDP',color = 'green',fontsize =15)
axes[0][2].set_title('EDUCATION GDP IN MILLONS',color = 'blue',fontsize =15)

for label in axes[0][2].get_xticklabels():
    label.set_color('crimson')
for label in axes [0][2].get_xticklabels():
    label.set_color('crimson')  

axes[1][0].step(gdp['Year'],gdp['CrudeOil'],color = 'blue')
axes[1][0].set_xlabel('Year',color = 'yellow',fontsize =15)
axes[1][0].set_ylabel('GDP',color = 'yellow',fontsize =15)
axes[1][0].set_title('CRUDEOIL GDP IN MILLONS',color = 'red',fontsize =15)

for label in axes[1][0].get_xticklabels():
    label.set_color('crimson')
for label in axes [1][0].get_xticklabels():
    label.set_color('crimson')  

axes[1][1].bar(gdp['Year'],gdp['ICT'],color = 'green')
axes[1][1].set_xlabel('Year',color = 'cyan',fontsize =15)
axes[1][1].set_ylabel('GDP',color = 'cyan',fontsize =15)
axes[1][1].set_title('ICT GDP IN MILLONS',color = 'blue',fontsize =15)

for label in axes[1][1].get_xticklabels():
    label.set_color('crimson')
for label in axes [1][1].get_xticklabels():
    label.set_color('crimson')  

axes[1][2].step(gdp['Year'],gdp['CropProduction'],color = 'purple')
axes[1][2].set_xlabel('Year',color = 'red',fontsize =15)
axes[1][2].set_ylabel('GDP',color = 'red',fontsize =15)
axes[1][2].set_title('CROP-PRODUCTION IN MILLONS',color = 'red',fontsize =15)

for label in axes[1][2].get_xticklabels():
    label.set_color('crimson')
for label in axes [1][2].get_xticklabels():
    label.set_color('crimson')  
```
